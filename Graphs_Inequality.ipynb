```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
filename = ('/content/_Viz5_Feb_2021_data - Perceived Obstacles to Gender Equality Focus 2030  Women Deliver .xlsx')
df = pd.read_excel(filename)
```

```python
df.head()
```

```python
filter = df['Country'] == 'United States'
newdf_2 = df[filter]
newdf = newdf_2.groupby('Response')['% of respondents'].mean().sort_values().reset_index(name = '% Response for Each Question')
newdf
newdf.head()
```

```python
men_filter = df['Gender'] == 'Male'
women_filter = df['Gender'] == 'Female'
men_of_america = df[filter & men_filter]
women_of_america = df[filter & women_filter]
new_men_df = men_of_america.groupby('Response')['% of respondents'].mean().sort_values().reset_index(name = '% Response for Each Question')
new_women_df = women_of_america.groupby('Response')['% of respondents'].mean().sort_values().reset_index(name = '% Response for Each Question')
```

```python
plt.style.use('seaborn-poster')
plt.barh(new_men_df['Response'], new_men_df['% Response for Each Question'],
         color = ((.5,.5,1), (.5,.5,1),(0,.5,1),(0,.4,1),(0,.3,1),(0,.2,1),(0,.1,1),(0,0,1), (0,0,1), (0.4,1,0.7), (0,1,0)));
plt.title('Reasons for Gender Inequality as perceived by Men in the US\nMen and Women have Similar Perceptions About Reasons for Inequality')
plt.xlabel('Percentage')
plt.xlim(0,.3)
colors = {'#1: boys and girls treated differently':(0.4,1,0.7), '#11: crises have unequal effects' : (.5,.5,1)}         
labels = list(colors.keys())
handles = [plt.Rectangle((0,0),1,1, color=colors[label]) for label in labels]
plt.legend(handles, labels);
```

```python
plt.style.use('seaborn-poster')
plt.barh(new_women_df['Response'], new_women_df['% Response for Each Question'],
         color = ((.5,.5,1), (.5,.5,1),(0,.5,1),(0,.4,1),(0,.3,1),(0,.2,1),(0,.1,1),(0,0,1), (0,0,1), (0.4,1,0.7), (0,1,0)));
plt.title('Reasons for Gender Inequality as perceived by Women in the US\nMen and Women have Similar Perceptions About Reasons for Inequality')
plt.xlabel('Percentage')
colors = {'Same #1 Reason as Men':(0,1,0), 'Same Top 3 (#2 and #3 switched)' : (0.4,1,0.7), '"different levels of bodily control" 3 places higher':(0,.1,1), 'Same Bottom 2 Reasons as Men' : (.5,.5,1)}         
labels = list(colors.keys())
handles = [plt.Rectangle((0,0),1,1, color=colors[label]) for label in labels]
plt.legend(handles, labels);
```
